### Import Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

### Access Web

```python
url = "https://shriyani.com/collections/all-products"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

### Web Site Title

```python
title = soup.title
print(title)
```

### Extract Data

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
rows = soup.find_all('div')
rows
```

```python
cards = soup.select("div.card.card--standard")
cards
```

### Create List for Store Data

```python
list_rows = [] 
```

### Extract Data From Elements

```python


for card in cards:


    name_tag = card.select_one("h3.card__heading a")
    name = name_tag.get_text(strip=True) if name_tag else None


    url = name_tag.get("href") if name_tag else None


    price_tag = card.select_one(".price-item--regular")
    price = price_tag.get_text(strip=True) if price_tag else None


    rating_tag = card.select_one(".rating-badge-value")
    rating = rating_tag.get_text(strip=True) if rating_tag else None


    size_tags = card.select(".size-option:not(.size-option--unavailable)")
    sizes = [s.get_text(strip=True) for s in size_tags]


    img_tag = card.select_one(".card__media img")
    image = None
    if img_tag:
        src = img_tag.get("src")
        image = "https:" + src if src.startswith("//") else src

    row = {
        "Name": name,
        "URL": url,
        "Price": price,
        "Rating": rating,
        "Sizes": ", ".join(sizes),  
        "Image": image
    }


    list_rows.append(row)
```

### Convert to Data Frame

```python
df = pd.DataFrame(list_rows)
print(df.head())
```

### Export as Excel

```python
df.to_excel("products.xlsx", index=False)
```
