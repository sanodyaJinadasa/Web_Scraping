```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
from bs4 import BeautifulSoup
%matplotlib inline
```

```python
url = "https://www.etsy.com/listing/226173957/rabbit-sweatshirt-cozy-fleece-bunny?external=1&ref=hp_consolidated_gifting_listings-2&pro=1&sts=1&logging_key=dca623c1abd3df1cd157cae049cbe8d5b3ebcebb%3A226173957"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
title = soup.title
print(title)
```

```python
soup = BeautifulSoup(html, 'lxml')
reviews = soup.select("div.common-reviews-new__list-item")
```

```python
all_links = soup.find_all("div")
for link in all_links:
    print(link.get("href"))
```

```python
data = []

for review in reviews:
    
    # Username
    username = review.select_one("div.nikename")
    username = username.get_text(strip=True) if username else None
    
    # Date
    date = review.select_one("div.date")
    date = date.get_text(strip=True) if date else None
    
    # Rating
    rating = review.select_one("div.rate-star")
    if rating and rating.has_attr("aria-label"):
        rating_value = rating["aria-label"].replace("Rating", "")
    else:
        rating_value = None
    
    # Review Text
    text = review.select_one("div.rate-des")
    text = text.get_text(strip=True) if text else None
    
    # Color and Size
    fit_items = review.select("span.rate-fit__item")
    
    color = None
    size = None
    
    for item in fit_items:
        label = item.get_text(strip=True)
        if "Color:" in label:
            color = label.replace("Color:", "").strip()
        if "Size:" in label:
            size = label.replace("Size:", "").strip()
    
    # Helpful Count
    helpful = review.select_one("span.like-num")
    helpful = helpful.get_text(strip=True) if helpful else None
    
    data.append({
        "Username": username,
        "Date": date,
        "Rating": rating_value,
        "Color": color,
        "Size": size,
        "Review": text,
        "Helpful": helpful
    })

df = pd.DataFrame(data)
print(df.head())
```

```python
df.to_excel("reviews_product.xlsx", index=False)
```
