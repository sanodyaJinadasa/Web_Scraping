```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
from bs4 import BeautifulSoup
%matplotlib inline
```

```python
url = "https://www.aliexpress.com/item/1005006729827689.html?spm=a2g0o.tm1000029706.d0.1.629f474cngf84m&sourceType=561&pvid=1686ae7c-a4f1-4fe6-a642-4670ca526323&pdp_ext_f=%7B%22ship_from%22:%22CN%22,%22sku_id%22:%2212000038113998217%22%7D&scm=1007.28480.478283.0&scm-url=1007.28480.478283.0&scm_id=1007.28480.478283.0&aecmd=true"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
title = soup.title
print(title)
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
list_rows = [] 
```

```python
reviews = soup.find_all("div", class_="list--itemReview--d9Z9Z5Z")
for review in reviews:
    print(review.text.strip())
    list_rows.append(review.text.strip())
```

```python
df = pd.DataFrame(list_rows)
print(df.head())
```
