```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from urllib.request import urlopen
from bs4 import BeautifulSoup
%matplotlib inline
```

```python
url = "https://www.trustpilot.com/review/www.shopify.com"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
title = soup.title
print(title)
```

```python
all_links = soup.find_all("a")
for link in all_links:
    print(link.get("href"))
```

```python
cards = soup.select("p.CDS_Typography_appearance-default__68c681.CDS_Typography_prettyStyle__68c681.CDS_Typography_body-l__68c681")
cards
```

```python
list_rows = [] 
```

```python
for card in cards:
    print(card.text.strip())
    list_rows.append(card.text.strip())
```

```python
df = pd.DataFrame(list_rows)
print(df.head())
```

```python
df.to_excel("customer_reviews.xlsx", index=False)
```
